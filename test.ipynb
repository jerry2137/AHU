```python
def remove_outliers(col):
    if ('Temperature' in col.name) | ('Air' in col.name):
        mean = col.mean()
        sd = col.std()
        return col.clip(mean-(3*sd), mean+(3*sd))
    if 'On/Off' in col.name:
        return col.clip(upper=1)
    if 'Freq' in col.name:
        return col.clip(upper=50)
    if 'Cooling Valve' in col.name:
        return col.clip(upper=100)
    

df = df.clip(lower=0).apply(remove_outliers)
```

```python
def summarise(df, keyword):

    title = df.columns.str.split(' ', n=1).str[0][0]
    df.columns = df.columns.str.split(' ', n=1).str[1]
    df = df.rename(columns=realname)
    # print(title)

    # all_names.update(df.columns.tolist())

    def merge(df):
        return_df = pd.Series(index=['Supply Air Temperature', 'Return Air Temperature', 'Return Temperature Setpoint'], dtype='float64')
        if not 'On/Off Status' in df:
            df['On/Off Status'] = 1
        if (df['On/Off Status'] == 0).all():
            return_df[:] = 'OFF'
        elif (df['On/Off Status'].mean() == 0) | df['On/Off Status'].isna().all():
            return_df[:] = 'N/A'
        else:
            wanted_columns = ['Supply Temperature', 'Return Temperature', 'Return Temperature Setpoint']
            missing_columns = list(set(wanted_columns) - set(df.columns.to_list()))
            return_df = df.loc[df['On/Off Status'] > 0, df.columns.intersection(wanted_columns)].mean()
            for missing_column in missing_columns:
                return_df[missing_column] = 'N/A'
            return_df = return_df[wanted_columns]
        return return_df

    df = df.groupby(by=[df.index.year, df.index.month]).apply(merge)
    
    df.index = df.index.map(lambda idx: f'{idx[0]}/{idx[1]}')

    return df.stack(0)

# all_names = set()
for df, keyword in [(df_AHU, 'AHU')]:
    df.groupby(df.columns.str.split(' ', n=1).str[0], axis=1).apply(summarise, (keyword))
# all_names

# summary_df = df.groupby(df.columns.str.rsplit('-', n=1).str[0].str.replace('-AHU', ''), axis=1).apply(summarise)
# summary_df.to_excel('summary_transformed.xlsx')
# summary_df
```
